```python
#!/bin/python3
import matplotlib.pyplot as plt
import pandas as pd
import math
import numpy as np
import copy
import csv
from matplotlib import pyplot as plt
file = "iso_geo_corr_loop_c_2198_800_0170_0001.txt"
d = []
with open(file) as f:
    reader = csv.reader(f, delimiter="\t")
    for number in reader:
        d.append(number)
        print(type(number))
print(type(d))
foo = [item for inner in d for item in inner]
#print(d)
print(foo)
print(foo[20])
```

```python
from matplotlib import pyplot as plt
file = "iso_geo_corr_loop_c_2198_400_0183_0001.txt"
d = []
with open(file) as f:
    reader = csv.reader(f, delimiter="\t")
    for number in reader:
        d.append(number)
        print(type(number))
print(type(d))
foo = [item for inner in d for item in inner]
index = [i for i in range(0,len(foo))]
arr = np.array(foo)
arr_fl = np.asarray(arr, dtype=float)
np_ind = np.array(index)
plt.title("Line Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.plot(np_ind, arr_fl, color="red")
plt.show()
```
