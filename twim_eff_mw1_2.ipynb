```python
#!/bin/python3
import matplotlib.pyplot as plt
import pandas as pd
import math
import numpy as np
import copy
```

```python
filename = "total_eff_twim.txt"
df = pd.read_csv(filename,sep='\t',names=["datatype","targettype","RunID","energy","twim_eff"])

## plot efficiencies for thin/medium/thick/empty runs
eff_thin = []
eff_medium = []
eff_thick = []
eff_empty = []
energy_vals = ["400","550","650","800"]
for ene in energy_vals:
    print(float(df.loc[(df["targettype"] == "c_54") & (df["energy"] == float(ene))]['twim_eff']))
    eff_thin.append(float(df.loc[(df["targettype"] == "c_54") & (df["energy"] == float(ene))]['twim_eff']))
    eff_medium.append(float(df.loc[(df["targettype"] == "c_1086") & (df["energy"] == float(ene))]['twim_eff']))
    eff_thick.append(float(df.loc[(df["targettype"] == "c_2198") & (df["energy"] == float(ene))]['twim_eff']))
    eff_empty.append(float(df.loc[(df["targettype"] == "empty") & (df["energy"] == float(ene))]['twim_eff']))
    
fig, ax = plt.subplots(1,3,figsize=(15, 6))
fig.tight_layout()
print(eff_thin)

##thin vs empty
ax[0].scatter(energy_vals,eff_thin,label="thin",color="blue",marker="s")
ax[0].scatter(energy_vals,eff_empty,label="empty",color="black",marker="s")
ax[0].legend()
ax[0].set_ylim(0.994,1)
ax[0].grid()
##medium vs empty
ax[1].scatter(energy_vals,eff_medium,label="medium",color="green",marker="v")
ax[1].scatter(energy_vals,eff_empty,label="empty",color="black",marker="s")
ax[1].legend()
ax[1].set_ylim(0.994,1)
ax[1].grid()
##thick vs empty
ax[2].scatter(energy_vals,eff_thick,label="thick",color="red",marker="o")
ax[2].scatter(energy_vals,eff_empty,label="empty",color="black",marker="s")
ax[2].legend()
ax[2].set_ylim(0.994,1)
ax[2].grid()
plt.show()
```

```python
print(eff_empty)
print("next'")

eff_corr_thin = [x/y for x,y in zip(eff_empty,eff_thin) ]
eff_corr_medium = [x/y for x,y in zip(eff_empty,eff_medium) ]
eff_corr_thick = [x/y for x,y in zip(eff_empty,eff_thick) ]
print(eff_corr_thin)

fig, ax = plt.subplots(figsize=(10, 8))
fig.tight_layout()
ax.scatter(energy_vals,eff_corr_thin,label="thin",color="blue",marker="s")
ax.scatter(energy_vals,eff_corr_medium,label="medium",color="green",marker="v")
ax.scatter(energy_vals,eff_corr_thick,label="thick",color="red",marker="o")
ax.legend()
ax.grid()

```
