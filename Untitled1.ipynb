```python
#!/bin/python3
import matplotlib.pyplot as plt
import pandas as pd
import math
import numpy as np
import copy
print(plt.style.available)
```

```python
filename = "all_params_cut_mw0_r3bmusic.txt"
df = pd.read_csv(filename,sep='\t',names=["datatype","targettype","RunID","energy","min_mw0_x","max_mw0_x","min_mw0_y","max_mw0_y","min_charge_r3b","max_charge_r3b"])
df["mean_mw0_x"] = (df["max_mw0_x"]+df["min_mw0_x"])/2
#df["mw0_err_x"] = abs((df["max_mw0_x"]-df["min_mw0_x"])/2)

df["mw0_err_x"] = abs((df["max_mw0_x"]-df["min_mw0_x"]))   #2sigma

df["mean_mw0_y"] = (df["max_mw0_y"]+df["min_mw0_y"])/2
#df["mw0_err_y"] = abs((df["max_mw0_y"]-df["min_mw0_y"])/2)

df["mw0_err_y"] = abs((df["max_mw0_y"]-df["min_mw0_y"]))   #2sigma

df["r3b_mean_charge"] = (df["max_charge_r3b"] + df["min_charge_r3b"])/2
df["r3b_err_charge"] = abs((df["max_charge_r3b"] - df["min_charge_r3b"])/2)
energy_vals = ["400","550","650","800"]
#for ene in energy_vals:
df.loc[df["targettype"] == "c_2198","energy"] = df.loc[df["targettype"] == "c_2198"]["energy"] -10
df.loc[df["targettype"] == "c_1086","energy"] = df.loc[df["targettype"] == "c_1086"]["energy"] -5
df.loc[df["targettype"] == "empty","energy"] = df.loc[df["targettype"] == "empty"]["energy"] + 5
fig, ax = plt.subplots(1,2)
fig.set_figheight(8)
fig.set_figwidth(14)
ax[0].errorbar(df.loc[df["targettype"] == "c_2198"]["energy"],df.loc[df["targettype"] == "c_2198"]["mean_mw0_x"],yerr = df.loc[df["targettype"] == "c_2198"]["mw0_err_x"],color='red',fmt="o",label="thick target")
ax[0].errorbar(df.loc[df["targettype"] == "c_1086"]["energy"],df.loc[df["targettype"] == "c_1086"]["mean_mw0_x"],yerr = df.loc[df["targettype"] == "c_1086"]["mw0_err_x"],color='green',fmt="o",label="medium target")
ax[0].errorbar(df.loc[df["targettype"] == "c_54"]["energy"],df.loc[df["targettype"] == "c_54"]["mean_mw0_x"],yerr = df.loc[df["targettype"] == "c_54"]["mw0_err_x"],color='blue',fmt="o",label="thin target")
ax[0].errorbar(df.loc[df["targettype"] == "empty"]["energy"],df.loc[df["targettype"] == "empty"]["mean_mw0_x"],yerr = df.loc[df["targettype"] == "empty"]["mw0_err_x"],color='black',fmt="o",label="empty")

ax[0].legend(fontsize=13)
ax[0].grid()
ax[0].set_xlabel("Beam Energy [AMeV]",fontsize=15)
ax[0].set_ylabel("X-Position on MWPC0 [mm]",fontsize=15)
ax[0].tick_params(axis='both', which='major', labelsize=15)


ax[1].errorbar(df.loc[df["targettype"] == "c_2198"]["energy"],df.loc[df["targettype"] == "c_2198"]["mean_mw0_y"],yerr = df.loc[df["targettype"] == "c_2198"]["mw0_err_y"],color='red',fmt="o",label="thick target")
ax[1].errorbar(df.loc[df["targettype"] == "c_1086"]["energy"],df.loc[df["targettype"] == "c_1086"]["mean_mw0_y"],yerr = df.loc[df["targettype"] == "c_1086"]["mw0_err_y"],color='green',fmt="o",label="medium target")
ax[1].errorbar(df.loc[df["targettype"] == "c_54"]["energy"],df.loc[df["targettype"] == "c_54"]["mean_mw0_y"],yerr = df.loc[df["targettype"] == "c_54"]["mw0_err_y"],color='blue',fmt="o",label="thin target")
ax[1].errorbar(df.loc[df["targettype"] == "empty"]["energy"],df.loc[df["targettype"] == "empty"]["mean_mw0_y"],yerr = df.loc[df["targettype"] == "empty"]["mw0_err_y"],color='black',fmt="o",label="empty")
ax[1].tick_params(axis='both', which='major', labelsize=15)
ax[1].legend(fontsize=13)
ax[1].grid()
ax[1].set_xlabel("Beam Energy [AMeV]",fontsize=13)
ax[1].set_ylabel("Y-Position on MWPC0 [mm]",fontsize=15)

fig.suptitle('MWPC0 X-Y Position Cuts', fontsize=18)
#ax.set_title("MWPC0 X seletion cuts")
fig.savefig("mwpc0_cutxy.png",dpi=300)
    
```
